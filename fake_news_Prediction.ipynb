## **Importing Libraries**

```python
import numpy as np
import pandas as pd
```

## **Inspecting Data**

```python
df_true=pd.read_csv("/content/True.csv",error_bad_lines=False)
```

```python
df_fake=pd.read_csv("/content/Fake (2).csv")
```

```python
df_true
```

```python
df_fake
```

```python
df_true["label"]=1
df_fake["label"]=0
```

```python
df_merge=pd.concat([df_true,df_fake],axis=0)
df_merge
```

```python
df_merge['label'].value_counts()
```

```python
df_merge.drop(['title','subject','date'],axis=1,inplace=True)
```

```python
# shuffling the dataset
df_merge.sample(frac=1)
```

```python
df_merge.isna().sum()
```

```python
df_merge.dropna(inplace=True)
```

```python
df_merge.isna().sum()
```

## **Text Data Cleaning**

```python
import re
# Define a function to remove special characters from a text
def remove_special_characters(text):
    # Use re.sub to remove special characters

    cleaned_text = re.sub(r'[^A-Za-z0-9\s]', '', text)
    cleaned_text = cleaned_text.lower()
    return cleaned_text

# Apply the function to the text column
df_merge['text'] = df_merge['text'].apply(remove_special_characters)

# Print the modified DataFrame
print(df_merge)
```

### Saperating features and target variables

```python
x=df_merge['text']
y=df_merge['label']
```

```python
x
```

## **Splitting the dataset into train and test**

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets with a test size of 0.5 (50%)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5)
```

## Text Data Vectorization using TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create a TF-IDF vectorizer with specific settings
tfidf = TfidfVectorizer(stop_words='english', max_df=0.8)

# Fit and transform the training data
x_train_tfidf = tfidf.fit_transform(x_train)

# Transform the testing data using the same vectorizer
x_test_tfidf = tfidf.transform(x_test)
```

```python
x_train_tfidf
```

## ***Machine Learning Algorithms ***

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR=LogisticRegression()
LR.fit(x_train_tfidf,y_train)
```

```python
pred_lr=LR.predict(x_test_tfidf)
```

```python
from sklearn.metrics import confusion_matrix
CM=confusion_matrix(y_test,pred_lr)
```

```python
CM
```

### Checking accuracy score

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,pred_lr)
acc
```

```python
# Import necessary libraries
from sklearn.metrics import precision_score

# Calculate precision
precision = precision_score(y_test,pred_lr)

# Print the precision score
print("Precision:", precision)
```

### **Decision Tree classifier**

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion="entropy",random_state=42)
classifier.fit(x_train_tfidf,y_train)
# criterion parameter is the function used to measure the quality of a split of tree nodes
# and it is  "gini" or "entropy"
# entropy is more complex since it makes use of algorithms so the calculation of the Gini Index will be faster
```

```python
y_pred=classifier.predict(x_test_tfidf)
```

```python
CM=confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,pred_lr)
acc
```

### Random Forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=10,criterion="entropy",random_state=42)
classifier.fit(x_train_tfidf,y_train)
y_pred=classifier.predict(x_test_tfidf)
from sklearn.metrics import confusion_matrix
CM=confusion_matrix(y_test,y_pred)
CM
```
